# SISTEMA DE ECUACIONES

$$2x_1 +3x_2 = 11$$
$$2x_1+10x_2=16$$

# DIVERGE

**Método Gauss-Jacobi**

```python
%autoreload 2
from src.iterative_methods import gauss_jacobi
import numpy as np
import matplotlib.pyplot as plt


A = np.array([[2, 3],
              [10, 2]], dtype=float)

b = np.array([16, 11], dtype=float)


x0 = np.array([[1], [1]])  # Initial guess (1, 1)
tol = 0.001  # Tolerance
max_iter = 100  # Maximum number of iterations

# Solve using Gauss-Jacobi method
x_jacobi, tray_jacobi = gauss_jacobi(A=A, b=b, x0=x0, tol=tol, max_iter=max_iter)
print("Solution using Gauss-Jacobi method:", x_jacobi)
```

```python
# Convertir la trayectoria obtenida a un array para x1 y x2
tray_jacobi_array = np.array(tray_jacobi)  # tray_jacobi contiene los valores de las iteraciones
x1_vals, x2_vals = tray_jacobi_array[:, 0], tray_jacobi_array[:, 1]

# Crear un gráfico para mostrar la divergencia
plt.figure(figsize=(10, 6))
plt.plot(x1_vals, x2_vals, label='Divergencia (x1 vs x2)', color='blue', marker='o', linestyle='-')

# Configurar el gráfico
plt.title("Divergencia de x1 y x2 - Método Gauss-Jacobi")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.grid(True)
plt.show()
```
