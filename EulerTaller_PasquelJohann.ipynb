## 1. Resolver la EDO 

Resolver la ecuación diferencial ordinaria:

$$
\frac{dy}{dt} = y - t^2 + 1
$$

en el intervalo: $0 \leq t \leq 2$ con la condición inicial:

$y(t_0) = 0.5$ 


```python
%load_ext autoreload
```

**N= 10**

```python
%autoreload 2
import pandas as pd
import matplotlib.pyplot as plt
from src.ODE import ODE_euler

def f(t, y):
    return y - t**2 + 1

# Parámetros del problema
a = 0        # Inicio del intervalo
b = 2        # Fin del intervalo
y_t0 = 0.5   # Condición inicial
N = 10       # Número de pasos

# Resolver la EDO
ys, ts, h = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=N)
df = pd.DataFrame({"i": range(N+1), "t_i": ts, "y_aprox": ys})
display(df)

plt.plot(ts, ys, marker="o", linestyle="-", color="b", label="N = 10")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Aproximación de la EDO usando Euler (N=10)")
plt.legend()
plt.grid()
plt.show()

```

**Real vs Aproximación**

**Y real**
$$y=(t+1)^2 -0.5e^t$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la solución exacta de la EDO
def y_real(t):
    return (t + 1)**2 - 0.5 * np.exp(t)

# Calcular los valores exactos en los mismos puntos de malla
y_exact = [y_real(t) for t in ts]

# Graficar la comparación
plt.figure(figsize=(8, 5))
plt.plot(ts, ys, marker="o", linestyle="-", color="b", label="N=10")
plt.plot(ts, y_exact, marker="s", linestyle="--", color="r", label="Solución Exacta")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Aproximación vs Y real")
plt.legend()
plt.grid()
plt.show()
```

## 3. Resolver la EDO

Resolver la ecuación diferencial ordinaria:

$$
\frac{dy}{dt} = \frac{y}{t} - (\frac{y}{t})^2
$$

en el intervalo: $1 \leq t \leq 2$ con la condición inicial:

$y(1) = 1$ 


**N=15**

```python
%autoreload 2
import pandas as pd
import matplotlib.pyplot as plt
from src.ODE import ODE_euler 


def f(t, y):
    return (y / t) - (y / t) ** 2

# Parámetros del problema
a = 1        # Nuevo inicio del intervalo
b = 2        # Fin del intervalo
y_t0 = 1     # Nueva condición inicial
N = 15       # Número de pasos


ys, ts, h = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=N)
df = pd.DataFrame({"i": range(N+1), "t_i": ts, "y_aprox": ys})
display(df)

# Graficar la solución aproximada
plt.plot(ts, ys, marker="o", linestyle="-", color="b", label="N = 10")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Aproximación de la EDO usando Euler (N=10)")
plt.legend()
plt.grid()
plt.show()
```

**Y real**
$$y=\frac{t}{1+ln(t)}$$

```python
%autoreload 2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from src.ODE import ODE_euler  # Importar el método de Euler

# Definir la ecuación diferencial dada
def f(t, y):
    return (y / t) - (y / t) ** 2

# Definir la solución exacta corregida
def y_real(t):
    return t / (1 + np.log(t))

# Parámetros del problema
a = 1        # Inicio del intervalo
b = 2        # Fin del intervalo
y_t0 = 1     # Condición inicial
N = 15       # Número de pasos

# Resolver la EDO usando el método de Euler
ys, ts, h = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=N)

# Calcular los valores exactos en los mismos puntos de malla
y_exact = [y_real(t) for t in ts]

# Crear DataFrame con número de iteración, t_i, y aproximaciones
df = pd.DataFrame({"i": range(N+1), "t_i": ts, "Euler (N=15)": ys, "Solución Exacta": y_exact})

# Mostrar la tabla con los valores
display(df)

# Graficar la comparación
plt.figure(figsize=(8, 5))
plt.plot(ts, ys, marker="o", linestyle="-", color="b", label="Aproximación (Euler, N=15)")
plt.plot(ts, y_exact, marker="s", linestyle="--", color="r", label="Solución Exacta")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Comparación: Método de Euler (N=15) vs Solución Exacta")
plt.legend()
plt.grid()
plt.show()
```
